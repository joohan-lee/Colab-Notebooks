```
import requests
from bs4 import BeautifulSoup

alib_result = requests.get("http://www.alba.co.kr/job/Main.asp?page=6&pagesize=50")
#print(alib_result.text)
alib_soup = BeautifulSoup(alib_result.text, "html.parser")
#print(alib_soup)
pagination = alib_soup.find("td",{"class":"title"})
pages=pagination.find_all('a')
print(pages)
```

```
def computepay(time, pay):
  if time > 40:
    total = 40 * pay + (time -40) * (pay*1.5)
  elif time <=40:
    total = time * pay
  else:
    print("unvalid")
    return -1
  return total

computepay(50,10)
```
